```python
contas = [1500.00, 800.00]

def mostrar_saldos():
    print(f"\nSaldos atuais:")
    print(f"1 - Conta Corrente : R$ {contas[0]:.2f}")
    print(f"2 - Poupança       : R$ {contas[1]:.2f}\n")

def validar_valor_positivo(valor_str):
    try:
        valor = float(valor_str.replace(',', '.'))
        if valor <= 0:
            print("O valor deve ser maior que zero.")
            return None
        return valor
    except:
        print("Valor inválido. Digite um número válido.")
        return None

def saque(indice_conta, valor):
    if contas[indice_conta] >= valor:
        contas[indice_conta] -= valor
        print(f"Saque de R$ {valor:.2f} realizado com sucesso.")
        return True
    else:
        print("Saldo insuficiente para saque.")
        return False

def transferencia(indice_origem, indice_destino, valor):
    if indice_origem == indice_destino:
        print("Origem e destino são a mesma conta.")
        return False
    if contas[indice_origem] >= valor:
        contas[indice_origem] -= valor
        contas[indice_destino] += valor
        print(f"Transferência de R$ {valor:.2f} realizada com sucesso.")
        return True
    else:
        print("Saldo insuficiente para transferência.")
        return False

def escolher_conta(mensagem):
    escolha = input(mensagem).strip()
    if escolha == '1':
        return 0
    elif escolha == '2':
        return 1
    else:
        print("Opção inválida.")
        return None

def menu():
    while True:
        print("=== Menu ===")
        print("1 - Mostrar saldos")
        print("2 - Sacar")
        print("3 - Transferir entre contas")
        print("0 - Sair")
        opc = input("Escolha uma opção: ").strip()

        if opc == '1':
            mostrar_saldos()

        elif opc == '2':
            mostrar_saldos()
            idx = escolher_conta("Escolha a conta (1-Corrente, 2-Poupança): ")
            if idx is None:
                continue
            v_str = input("Valor do saque: ").strip()
            valor = validar_valor_positivo(v_str)
            if valor is None:
                continue
            saque(idx, valor)
            mostrar_saldos()

        elif opc == '3':
            mostrar_saldos()
            origem = escolher_conta("Conta de origem (1-Corrente, 2-Poupança): ")
            if origem is None:
                continue
            destino = 1 - origem
            v_str = input("Valor a transferir: ").strip()
            valor = validar_valor_positivo(v_str)
            if valor is None:
                continue
            transferencia(origem, destino, valor)
            mostrar_saldos()

        elif opc == '0':
            print("Encerrando.")
            break

        else:
            print("Opção inválida.")

if __name__ == "__main__":
    menu()
```
